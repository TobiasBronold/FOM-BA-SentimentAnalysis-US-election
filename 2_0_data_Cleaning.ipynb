# Cleaning

First, the data is read in and randomly checked. The data is then cleansed using regex, e.g. for HTML tags in the comments.

# Imports

```python
import pandas as pd
import re
import csv
```

```python
csv_path = 'Export_csv/comments_v1.csv'
df_comments = pd.read_csv(
    csv_path, 
    encoding='utf-8', 
    quoting=csv.QUOTE_ALL,
    sep=',',  
    quotechar='"',
    engine='python' 
)
df_comments.info()
df_comments.head()
```

```python
# Funktion zum Entfernen von HTML-Tags
def clean_html_tags_in_column(series):
    def remove_html_tags(text):
        if isinstance(text, str):
            return re.sub(r'<[^>]*>', ' ', text)
        return text
    
    return series.apply(remove_html_tags)

df_comments['text'] = clean_html_tags_in_column(df_comments['text'])
print(df_comments)
```

```python
df_comments.to_csv('Export_csv/ImportantCSV/comments_v2.csv', encoding='utf-8', quoting=csv.QUOTE_ALL)
```
